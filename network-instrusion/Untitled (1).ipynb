```python
import time
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.impute import SimpleImputer
from imblearn.over_sampling import RandomOverSampler, SMOTE
from imblearn.under_sampling import RandomUnderSampler

# Assuming you have your feature matrix X and target vector y with missing values
# X, y = ...

# Impute missing values
imputer = SimpleImputer(strategy='mean')  # You can choose strategy based on your data
X_imputed = imputer.fit_transform(X)

# Define resampling models
resampling_models = {
    'Random OverSampler': RandomOverSampler(random_state=42),
    'SMOTE': SMOTE(random_state=42),
    'Random UnderSampler': RandomUnderSampler(random_state=42)
}

# Initialize lists to store computation times
model_names = []
computational_times = []

# Compute computation time for each model
for name, model in resampling_models.items():
    start_time = time.time()
    X_resampled, y_resampled = model.fit_resample(X_imputed, y)
    end_time = time.time()
    computation_time = end_time - start_time
    model_names.append(name)
    computational_times.append(computation_time)

# Compute percentage decrease in computational time compared to the first model
percentage_decrease = [(computational_times[0] - time) / computational_times[0] * 100 for time in computational_times]

# Plot the results
plt.figure(figsize=(10, 5))
plt.bar(model_names, percentage_decrease, color='skyblue')
plt.title('Percentage Decrease in Computational Time for Various Resampling Models')
plt.xlabel('Resampling Models')
plt.ylabel('Percentage Decrease in Time')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

from sklearn.svm import SVC
from sklearn.metrics import mean_squared_error,r2_score,accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
```

```python
support=SVC()
```

```python
support.fit(X_train, y_train)
```

```python
y_pred=support.predict(X_test)
```

```python
mse=mean_squared_error(y_pred,y_test)
r2=accuracy_score(y_pred,y_test)
print(mse)
print(r2)
```

```python
print(classification_report(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```
